```python
import matplotlib.pyplot as plt
x=[]
y=[]
for i in range (100):
    x.append(i**0.5)
    y.append(i**2)
plt.xlabel('X') 
plt.ylabel('Y')
plt.title('First test') 
plt.plot(x, y,color='red')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=[]
x = np.linspace(0, 2*np.pi, 100)
plt.xlabel('X') 
plt.ylabel('Y')
plt.title('Sec test') 
plt.plot(x, np.sin(x), color='Green')
plt.plot(x, np.cos(x), color='Red')
plt.show()
```
