```python
import pandas as pd # Для выгрузки этого
listalf=["а","б","в","г","д","е","ж","з","и","й","к","л"," м","н","о","п","р","с","т","у","ф","х","ц","ч","ш","щ","ъ","ы"," ь"," э"," ю","я"]
listalfval=[0 for i in range(len(listalf))]
fraza="волккаждыйгодлиняетзптавсесербывает."
for i in fraza:
    for j in listalf:
        if i in j:
            listalfval[listalf.index(i)] += 1
col_alf = pd.DataFrame({'Name':listalf, 'Col':listalfval})
col_alf.plot(kind="bar", y="Col", x="Name", rot=6, fontsize=10, figsize=(7,3))
```
