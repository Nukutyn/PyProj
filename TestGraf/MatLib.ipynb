```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random as rand
x=[]
y=[]
for i in range (100):
    x.append(i**0.5)
    y.append(i**2)
plt.xlabel('X') #название оси 
plt.ylabel('Y')
plt.title('First test') 
plt.plot(x, y,color='red')#сам график через matplotlib
plt.show()
```

```python
x=[]
x = np.linspace(0, 2*np.pi, 100)#генерации последовательности от 0 до пи с 100 точками в последовательности 
plt.xlabel('X') 
plt.ylabel('Y')
plt.title('Sec test') 
plt.plot(x, np.sin(x), color='Green')#тут синус из numpy
plt.plot(x, np.cos(x), color='Red')#а тут косинус из numpy
plt.show()
```

```python
stolbx =  [1,2,3,4,5,6,7]
stolby = []
for i in range(len(stolbx)):
    stolby.append(rand.randint(1, 50))#рандомнвая генерация инт значений
plt.title("Trird Test") 
plt.bar(stolbx,stolby)
plt.show()
```

```python
PieVal = []
PieName = ["Alfa","Beta","Gamma","Delta","Epsilon"]
for i in range(5):
    PieVal.append(rand.randint(15,100))
plt.title("Fourth Test") 
plt.pie(PieVal,labels=PieName)# диаграмма пирог через matplotlib также рандомная
plt.show()
```

```python
import pandas as pd
import seaborn as sns
arr = pd.read_csv('XandY.txt', encoding='UTF-8', sep='|')# выгрузка информации используя pandas
#arr.dtypes проверка типов хр и ур были object далее привожу их к флот
arr.xp=arr.xp.str.replace(',','.')# замена запятой на точку так как через нее не работало 
arr.yp=arr.yp.str.replace(',','.')
arr['xp']=pd.to_numeric(arr['xp'], errors='coerce') #приводим к числовому значению и перезаписываем 
arr['yp']=pd.to_numeric(arr['yp'], errors='coerce') 
```

```python
plt.scatter(arr['x'],arr['y'])#график рассеивания и его методы записи 
plt.scatter(arr['x'],arr['xp'])
plt.scatter(arr.x,arr.yp)
```

```python
sns.scatterplot(data=arr, x="x", y="y")#график рассеивания только через seaborn
sns.scatterplot(data=arr, x="x", y="xp")
sns.scatterplot(data=arr, x="x", y="yp")
```

```python
CompanyTypeNum = attack['Organization Type'].value_counts()# Подсчитал колическто типов огранизаций 

LittleCompany = CompanyTypeNum[CompanyTypeNum < 10]# Индексация с помощью условий 
BigCompany = CompanyTypeNum[CompanyTypeNum >= 10]

LitSum=LittleCompany.sum()
LittleCompany=pd.Series(LitSum, index=["Other"])# создаем новый объект Series и даем ему индекс Other

PieDiag = pd.concat([BigCompany,LittleCompany])# Объединяем BigCompany и LittleCompany в PieDiag используя concat
PieDiag.plot(kind="pie")
```
